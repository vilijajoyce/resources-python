Helpful for organizing data.

```python
# Create dictionary. Key-value pairs.
person = {'firstname': 'vilija', 'lastname': 'joyce', 'office': 324, 'hair': 'blonde'}
```

```python
# Another way to create a dictionary
person = dict(firstname='vilija', lastname='joyce', office=324, hair='blonde')
```

```python
# Look up any value by the key.
person['firstname']
```

```python
person['office']
```

```python
# Look up all keys aka variable names of a dictionary.
person.keys()
```

```python
# Look up all values
person.values()
```

```python
# Look up keys and values
person.items()
```

## Sets
Manipulate and analyze data. Collection of objects.
Think of them as tables - includes union statements.

```python
# Create animal set - this is unordered (unlike in a list)
animal_set = set(['coyote', 'dog', 'bear', 'cat', 'elephant'])
animal_set
```

```python
# Another way to create a set
set_1 = {2, 4, 6, 8, 10}
set_2 = {42, 'vilija', (1, 2, 3), 3.14159}
print(type(set_1))
```

```python
# Full union of 2 sets                  
x1 = {'lopinavir', 'ritonavir', 'lamivudine'}
x2 = {'abacavir', 'lamivudine', 'ritonavir'}
x1 | x2
```

```python
# Full union method 2
x1.union(x2)
```

```python
# Inner union or inner join
x1.intersection(x2)
```

```python
# Inner union method 2
x1 & x2
```

```python
# In x1 but not x2
x1.difference(x2)
```

```python
# In x1 but not x2 method 2
x1-x2
```

```python
# Outer join (in either set x1 or set x2, but not both)
x1.symmetric_difference(x2)
```

```python
# Outer join (in either set x1 or set x2, but not both) method 2
x1 ^ x2
```

```python
# Anything not in common? (weird)
x1.isdisjoint(x2)
```

```python
# Add and remove elements from set
x1.add('nevirapine')
x2.remove('lamivudine')
x1
x2
```

```python
# If else statements
# Assign raining variable
raining = False

if raining:
    print("Take umbrella") # If true
else:
  print("Take sunglasses") # If false

print("Take wallet") # Regardless
```

```python
# Multiple if else conditions
weather = 'cloudy'

if weather == 'raining':
    print("Take umbrella")
elif weather == 'sunny':
    print("Take sunglasses")
elif weather == 'cloudy':
    print("Take sweater")
print("Take wallet")
```

```python
# Multiple conditions with a final option
weather = 'tornado'

if weather == 'raining':
    print("Take umbrella")
elif weather == 'sunny':
    print("Take sunglasses")
elif weather == 'cloudy':
    print("Take sweater")
else:
    print("I don't know what to do!")
print("Take wallet")
```

```python
# For loops - automate/repeat code over and over again
# Set x equal to a, then print. Then x equal to b, then print.
# for x in ['a','b']
#   print x

# Print list of numbers between 0 and 5 (i=0, i=1, i=2...)
for i in range(0,5):
  print(i)
```

```python
# Loop over sequential data
art_rx = ['abacavir', 'ritonavir', 'lamivudine', 'nevirapine', 'zidovudine']

for art in art_rx:
  print(art)
```

```python
# For each med, add the word art to the list
art_rx = ['abacavir', 'ritonavir', 'lamivudine', 'nevirapine', 'zidovudine']

for item in range(len(art_rx)):
  art_rx.append('art')

print(art_rx)
```

```python
# Iterate through strings
vilija = 'vilija'

for letter in vilija:
    print(letter)
```

```python
# Iterate through a dictionary - print keys aka variables and values
directory = {'firstname': 'vilija', 'lastname': 'joyce', 'hair': 'blonde'}

for key in directory:
    print(key + ': ' + directory[key])
```

```python
# Nested loops - a loop within another loop
# for [first iterating variable] in [outer loop]: # Outer loop
# [do something] # Optional
# for [second iterating variable] in [nested loop]: # Nested loop
#   [do something]

# For each number, print every letter
num_list = [1, 2, 3]
alpha_list = ['a', 'b', 'c']

for number in num_list:
    print(number)
    for letter in alpha_list:
        print(letter)
```

```python
# List of lists - patients and longitudinal systolic and diastolic bp values
pts_bp_3mos = [['A01', 'A02', 'B01'], [120, 152, 130], [75, 100, 90]]

for l in pts_bp_3mos:
    print(l)
```

```python
# Nested for loop - print out each of the items of the list 1 by 1
for list in pts_bp_3mos:
    for item in list:
        print(item)
```

```python
# Create 4 different equations, plug in values, and return answers 
# Answers are called a tuple (similar to a list)
def arithmetic(x,y):
  _sum = x+y
  _diff = x-y
  _prod = x*y
  _div = x/y
  return _sum, _diff, _prod, _div

arithmetic(7,3)
```
